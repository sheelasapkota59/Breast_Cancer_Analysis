```python
# Sample DataFrame
data = pd.DataFrame({
    "Age": [40, 39, 45, 26, 21],
    "Breast Quadrant_Lower inner": [True, False, False, True, False],
    "Breast Quadrant_Lower outer": [False, False, True, False, False],
    "Diagnosis Result": ["Benign", "Benign", "Malignant", "Benign", "Benign"],
    "Breast": ["Right", "Left", "Left", "Right", "Left"]
})

# Manually identify categorical columns and numeric columns
categorical_columns = ["Breast Quadrant_Lower inner", "Breast Quadrant_Lower outer", "Diagnosis Result", "Breast"]
numeric_columns = ["Age"]  # Specifically include numeric columns

le = LabelEncoder()
data[categorical_columns] = data[categorical_columns].apply(le.fit_transform)
data
```

```python
x = data[["Age","Breast Quadrant_Lower inner", "Breast Quadrant_Lower outer", "Breast"]]
y = data['Diagnosis Result']

sc = StandardScaler()
x_scaled = sc.fit_transform(x)

x_train, x_test , y_train , y_test = train_test_split(x_scaled, y , test_size = 0.3 , random_state = 42 )

#train multiple model
models = {
    "svm": SVC(kernel = "linear"),
    "Random Forest" : RandomForestClassifier(n_estimators = 100 , random_state = 42),
    "XGboost": XGBClassifier(eval_matric = "logloss"),
    "logistic Regression" : LogisticRegression()
}

# Fit models and evaluate
for name , model in models.items():
    model.fit(x_train , y_train)
    y_pred = model.predict(x_test)
    evaluate = accuracy_score(y_pred , y_test)
    print(f"{name} Evaluate : {evaluate:.4f}")
    
```

```python
#Compare Predictions for Different Sample Groups
#how age affect cancer?
Age_threshold = data["Age"].median()
age_column_index = data.columns.get_loc("Age")  # Get index of "Age" in DataFrame
age_column = x_test[:, age_column_index]
young_patients = x_test[age_column <= Age_threshold]
old_patients = x_test[age_column > Age_threshold]

if len(young_patients) > 0:
    young_pred = models["Random Forest"].predict(young_patients)
    young_malignant_rate = (young_pred.sum() / len(young_pred)) * 100
    print(f"Malignancy Rate in Young Patients: {young_malignant_rate:.2f}%")
else:
    print("No young patients in the dataset.")

if len(old_patients) > 0:
    old_pred = models["Random Forest"].predict(old_patients)
    old_malignant_rate = (old_pred.sum() / len(old_pred)) * 100
    print(f"Malignancy Rate in Old Patients: {old_malignant_rate:.2f}%")
else:
    print("No old patients in the dataset.")
```
